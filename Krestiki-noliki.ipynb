# Игра "Крестики-нолики"

## Приветствие

```python
def greeting():
    print(f'''Привет, {username1} и {username2}!
Добро пожаловать в игру "Крестики-нолики"!
{username1} играет «крестиками», а {username2} — «ноликами».
Цель игры - выстроить в ряд 3 своих фигуры 
по вертикали, горизонтали или диагонали.
Удачи!''')
```

## Игровое поле

```python
# Создаём пустое поле:
field = [[' '] * 3 for i in range(3)]


# Напишем функцию для вывода поля с координатами:
def display_field():
    n = 0
    print()
    print('    0    1    2')
    for f in field:
        print(n, f)
        n += 1
    print()
```

## Ход

```python
def turn(username):
    print(f'Ход игрока {username}:')
    i = int(input('Введите номер строки: '))
    j = int(input('Введите номер столбика: '))
    if field[i][j] == ' ':
        if username == username1:  # Первый игрок ходит крестиками
            field[i][j] = 'x'
        elif username == username2:  # Второй игрок ходит ноликами
            field[i][j] = 'o'
    else:
        print('Поле занято!')
        turn(username)
```

## Условия выигрыша

```python
# Запишем все выигрышные комбинации:
def winning_condition():
    return field[0][0] != ' ' and (
        field[0][0] == field[0][1] == field[0][2]
        or field[0][0] == field[1][0] == field[2][0] or
        field[0][0] == field[1][1] == field[2][2]) or field[1][1] != ' ' and (
            field[1][0] == field[1][1] == field[1][2]
            or field[0][1] == field[1][1] == field[2][1] or field[2][0] ==
            field[1][1] == field[0][2]) or field[2][2] != ' ' and (
                field[2][0] == field[2][1] == field[2][2]
                or field[0][2] == field[1][2] == field[2][2])
```

## Процесс игры

```python
def game():
    display_field()
    count = 0
    player = username1
    while True:
        turn(player)
        display_field()
        count += 1
        # Проверяем, есть ли выигрышная комбинация
        # Начинаем проверять с пятого хода:
        if count >= 5:
            if winning_condition():
                print(f'Выиграл/а {player}!')
                break
        # После 9-и ходов все поля заполнены
        # Если нет выигрышной комбинации, то результат - ничья
        if count == 9:
            print('Ничья!')
            break
        # Меняем игрока для следующего хода:
        if player == username1:
            player = username2
        else:
            player = username1
```

## Собственно игра

```python
username1 = input('Введите имя первого игрока: ')
username2 = input('Введите имя второго игрока: ')
greeting()

while True:
    field = [[' '] * 3 for i in range(3)]
    game()
    restart = input('Хотите сыграть ещё раз? да или нет: ').lower()
    if restart == 'да':
        continue
    else:
        print('Спасибо за игру!')
        break
```
